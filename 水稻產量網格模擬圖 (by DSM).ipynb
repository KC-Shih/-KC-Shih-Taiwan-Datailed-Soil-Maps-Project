```r
# ==================== #
# ===   20210421   === #

library(Imap)
library(dplyr)
library(ggplot2)

setwd("C:/Users/user/Desktop/1091/thesisdata") # desktop
```

```r
# ======= read file ======== #
TWfiveyields = read.csv("TWfiveyields.csv")
TWoneyieldsNaomit = read.csv("TWoneyieldsNaomit.csv")
final_Han546 = read.table("final_Han546.txt" , header = T)
```

```r
# ======= grid plots ======== #

# ===  1 KM === #
col = as.numeric(as.vector(TWoneyieldsNaomit$yield)) # create color interval


ggplot(TWoneyieldsNaomit, aes(x=lon,y=lat))+
  geom_point( aes(x=lon,y=lat, color = col), size = 0.9)+
  scale_colour_gradient(low="black", high = "green",name ="Yield (kg/ha)")+
  labs(x = "longitude", y = "Latitude")

# ===  5 KM === #
col = as.numeric(as.vector(TWfiveyields$yield))
ggplot(TWfiveyields, aes(x=lon,y=lat))+
  geom_point( aes(x=lon,y=lat, color = col))+
  scale_colour_gradient(low="black", high = "green",name ="Yield(kg/ha)")+
  labs(x = "longitude", y = "Latitude")
```

```r

# ============================= #
# === Taiwan yield boxplots === #
# ============================= #

par(mfrow=c(1,5),cex.axis=1.2)

TWoneyieldsNaomit$yield[TWoneyieldsNaomit$lat>24.5
                        ] %>% boxplot(main = "Northern Taiwan", 
                                      ylim=c(100,5500),cex.main=1.2,outpch = 20)

TWoneyieldsNaomit$yield[TWoneyieldsNaomit$lat>23.63 & TWoneyieldsNaomit$lat< 24.5
                         & TWoneyieldsNaomit$lon < 121.1
                         ] %>% boxplot(main = "Central Taiwan", 
                                       ylim=c(100,5500),cex.main=1.2,outpch = 20)

TWoneyieldsNaomit$yield[TWoneyieldsNaomit$lat<23.63
                        ] %>% boxplot(main = "Southern Taiwan", 
                                      ylim=c(100,5500),cex.main=1.2,outpch = 20)

TWoneyieldsNaomit$yield[TWoneyieldsNaomit$lat< 24.5 & TWoneyieldsNaomit$lon > 121.1
                        ] %>% boxplot(main = "Eastern Taiwan",
                                      ylim=c(100,5500),cex.main=1.2,outpch = 20)

final_Han546$yield %>%  boxplot(main = "LSD",
                                ylim=c(100,5500) ,cex.main=1.2, outpch = 20)
```
