```r
# install.packages("Imap")
rm(list = ls())
library(Imap)
library(dplyr)
library(ggplot2)

setwd("C:/Users/user/Desktop/1091/thesisdata") # desktop

final_Han546 = read.table("final_Han546.txt", header = T)
final_DSM    = read.table(file="yield(grid all).txt",header = T)
onekmgrid   = read.csv("onekmgrid.csv")



TY_D = which(final_DSM$lat > 24.589 & final_DSM$lat < 25.116 
             & final_DSM$lon > 120.988 &  final_DSM$lon < 121.480)

TY_H = which(final_Han546$Latitude > 24.589 & final_Han546$Latitude < 25.116 
             & final_Han546$Longitude > 120.988 &  final_Han546$Longitude < 121.480)

TY_DSM = final_DSM[TY_D,]
TY_Han = final_Han546[TY_H,]

L = 74052
# ===== 計算五公里網格點跟DSM有產量的網格點的距離
li1 = data.frame()
for(i in 1:L){
  for(j in 1:920){
    
    li1[i,j] = gdist(onekmgrid$lat[i],onekmgrid$lon[i],
                    final_DSM$lat[j],final_DSM$lon[j],units = "km")
  } 
  print(round(i/L,3))
}

# ===== 取出最近的四個點 
x = list()
heads = list()
d = list()
d4 = list()
st = Sys.time()
for(i in 1:L){
  li1[i,]  %>% order() -> x[[i]]  # order 從小到大位置
  heads[[i]] = x[[i]][1:4]  #前四的位置
  sort(li1[i,])[1:4] -> d4[[i]] # 前四的距離
  
  print(round(i/L,3))
}
Sys.time() - st  # Time difference of 24.70612 mins
# da = list()
# for(i in 1:L){
#   d4[,i] %>% as.numeric() -> da[[i]]
#   print(round(i/L,3))
# }
st = Sys.time()
d4 %>% unlist() %>% as.numeric() -> r
matrix(r, nrow = 4) -> rm
Sys.time() - st  #Time difference of 0.1057169 secs


# ===== 取出這四個點的產量
st = Sys.time()
yield41km = list()


for(i in 1:L){
  yield41km[[i]] = final_DSM$yield[heads[[i]]]
}
Sys.time() - st # Time difference of 0.1107039 secs

fy = c() # 放最終產量加權結果
# 計算加權產量
st = Sys.time()
for(i in 1:74052){
  pointd1[i] = ( 1/rm[,i][1]) / ( (1/ (rm[,i][1])) + (1/ (rm[,i][2])) + 
                                    (1/ (rm[,i][3])) + (1/ (rm[,i][4]))   )
  pointd2[i] = ( 1/rm[,i][2]) / ( (1/ (rm[,i][1])) + (1/ (rm[,i][2])) + 
                                    (1/ (rm[,i][3])) + (1/ (rm[,i][4]))   )  
  pointd3[i] = ( 1/rm[,i][3]) / ( (1/ (rm[,i][1])) + (1/ (rm[,i][2])) + 
                                    (1/ (rm[,i][3])) + (1/ (rm[,i][4]))   )  
  pointd4[i] = ( 1/rm[,i][4]) / ( (1/ (rm[,i][1])) + (1/ (rm[,i][2])) + 
                                    (1/ (rm[,i][3])) + (1/ (rm[,i][4]))   )  
  
  # 該點的最終產量，依序計算再跟onekmgrid  cbind 即可，就可以畫圖
  c(pointd1[i],pointd2[i],pointd3[i],pointd4[i] )%*% yield41km[[i]] -> fy[i]
  print(round(i/74052,3))
}
Sys.time() - st  # Time difference of 3.25131 secs



cbind(onekmgrid[1:L,] , fy) -> oneyield
# colnames(oneyield) = c("lon", "lat", "yield")
# 
# col = as.numeric(as.vector(oneyield$yield))
# 
# ggplot(oneyield, aes(x=lon,y=lat))+
#   geom_point( aes(x=lon,y=lat, color = col))+
#   scale_colour_gradient(low="red", high = "yellow",name ="Yield")+
#   labs(x = "longitude", y = "Latitude")

# ======================= #
#   對應到詳測土壤圖      #
# ======================= #

library(rgdal)
library(raster)
setwd("C:/Users/USER/Desktop/1082/SoilMap/DetailedSoilMap/")  # desktop

datDe = readOGR("DetailedSoilMapWGS84.shp", encoding="UTF-8")

# 取出有在polygon裡面的
st = Sys.time()
inout1 = over(
  SpatialPoints(oneyield[,c("lon","lat")],proj4string=CRS(projection(datDe))),
  as(datDe,"SpatialPolygons")
)
Sys.time() - st  # Time difference of 7.052428 secs

cbind(onekmgrid[1:L,] , fy, inout1)  -> TWoneyields
TWoneyieldsNaomit = na.omit(TWoneyields)
TWoneyieldsNaomit$inout1 = TWoneyieldsNaomit$fy
colnames(TWoneyieldsNaomit) = c("lon", "lat", "yield", "yield.r")


col = as.numeric(as.vector(TWoneyieldsNaomit$yield)) #建立顏色區間


ggplot(TWoneyieldsNaomit, aes(x=lon,y=lat))+
  geom_point( aes(x=lon,y=lat, color = col), size = 0.5)+
  scale_colour_gradient(low="red", high = "yellow",name ="Yield")+
  labs(x = "longitude", y = "Latitude")

ggplot(TWoneyieldsNaomit, aes(x=lon,y=lat))+
  geom_point( aes(x=lon,y=lat, color = col), size = 0.01)+
  scale_colour_gradient(low="black", high = "green",name ="Yield")+
  labs(x = "longitude", y = "Latitude")

```
