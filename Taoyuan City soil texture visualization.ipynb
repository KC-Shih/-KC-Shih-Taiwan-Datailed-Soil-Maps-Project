```r
setwd("C:/Users/USER/Desktop/1082/SoilMap/DetailedSoilMap/")  # 桌電
library(sf)
library(RColorBrewer)
```

```r
datDe_sf = st_read("DetailedSoilMapWGS84.shp")
datDe_sf_T = datDe_sf[st_is_valid(datDe_sf)==TRUE,]

# datDe_sf_T$第一層質地 = as.factor(datDe_sf_T$第一層質地)

# plot(datDe_sf_T["第一層質地"],axes=T)

```

```r
TY = datDe_sf_T[datDe_sf_T$COUNTYNAME == "桃園縣",] 
TY = na.omit(TY)

TY$第一層質地 = as.factor(TY$第一層質地)
TY$第二層質地 = as.factor(TY$第二層質地)
TY$第三層質地 = as.factor(TY$第三層質地)
TY$第四層質地 = as.factor(TY$第四層質地)
```

```r
plot(TY["第一層質地"],axes=T, main = "Taoyuan 1st profile soil texture",pal = brewer.pal(n = 9, name = "Paired"),)

plot(TY["第二層質地"],axes=T, main = "Taoyuan 2nd profile soil texture",pal = brewer.pal(n = 9, name = "Paired"),)

plot(TY["第三層質地"],axes=T, main = "Taoyuan 3rd profile soil texture",pal = brewer.pal(n = 9, name = "Paired"),)

plot(TY["第四層質地"],axes=T, main = "Taoyuan 4th profile soil texture",pal = brewer.pal(n = 9, name = "Paired"),)
```
